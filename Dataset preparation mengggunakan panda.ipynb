1. Masukkan iris.csv ke dataframe

```python
import pandas as pd
```

```python
iris = pd.read_csv('iris.csv')
iris
```

Penjelasan : Kita importkan terlebih dahulu pandas dengan mengaliaskan pd agar mudah untuk penyebutannya. Lalu kita bisa membaca file iris.csv nya dengan menggunakan perintah pd.read_csv yang ditampung dalam variabel iris. Lalu untuk memanggilnya kita bisa memakai variabel iris. Lalu ketika dirun akan menampilkan dataset iris yang berisi sepal.legth, sepal.width, petal.length, petal.width, dan variety. 

2. Normalisasikan setiap attribut iris ke dataframe menggunakan normalisasi min max sehingga
setiap nilai attribut memiliki range 0 – 1. Vbaru = (Vlama - Vmin) / (Vmax - Vmin). Vbaru = nilai
data di suatu sel setelah normalisasi, Vlama = nilai data asli sebelum normalisasi, Vmax = nilai
max dari kolom data tersebut, Vmin = nilai min dari kolom data tersebut 

```python
list_iris = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

for data in list_iris:
  vmin = iris[data].min()
  vmax = iris[data].max()

  iris[data] = (iris[data] - vmin) / (vmax - vmin)

iris
```

Penjelasan : pertama tama kita ambil atribut yang dimasukkan ke variabel list_iris. Lalu untuk normalisasinya, kita pakai perulangan for data in list_iris. lalu untuk mencari vmin dan vmax kita pakai index variabel iris dengan index data untuk mencarinya. Setelah mendapatkan hasil vmin dan vmax, kita terapkan rumus Vbaru = (Vlama - Vmin) / (Vmax - Vmin). Lalu untuk mengecek apakah sudah dinormalisasi atau belum, kita bisa memanggil variabel iris untuk menampilkannya. 

3. Cari centroid dari data yang telah dinormalisasi. Centroid dihitung dari rata2 seluruh data suatu 
kolom

```python
cluster_iris = iris['variety'].unique()

centroid = dict()

list_iris = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

for cluster in cluster_iris:
  new_iris = iris[iris['variety'] == cluster]
  centroid[cluster]={}
  for data in list_iris:
    centroid[cluster][data] = new_iris[data].mean()
    
centroid
```

Penjelasan : awal kita memakai iris dengan index variety dan ditampung dalam variabel cluster_iris. Selanjutnya kita buat variabel centroid yang berisi dictionary kosong lalu nanti hasil centroid itu akan ditampung dalam variabel centroid tersebut. Lalu kita ambil setiap atribut yang ada di iris dan dimasukkan dalam variabel list_iris. Lalu kita masuk ke perulangan for cluster in cluster_iris, lalu jika iris variety nya sama dengan cluster, maka akan dimasukkan ke variabel new_iris. Kemudian kita deklarasikan centeroid dengan dictionary kosong. Lalu kita tambahkan lagi perulangan dalam perulangan sebelumnya yaitu for data in list_iris lalu kita cari rata rata dari variabel new_iris dengan index data lalu dimasukkan ke variabel centroid[cluster][data]. Lalu untuk menunjukkan hasilnya kita bisa memanggil variabel centroid tadi. 

4. Masukkan iristesting.csv ke dataframe. Normalisasikan iristesting.csv menggunakan vmin dan 
vmax pada data 1

```python
iristesting = pd.read_csv('iristesting.csv')

list_iris = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

for data in list_iris:
  vmin = iristesting[data].min()
  vmax = iristesting[data].max()

  iristesting[data] = (iristesting[data] - vmin) / (vmax - vmin)

iristesting
```

Penjelasan : Pertama, seperti tadi untuk membaca file iristesting kita bisa menggunakan perintah pd.read_csv dan dimasukkan ke variabel iristesting. Lalu kita ambil atribut dari iristesting kemudian kita masukkan ke variabel list_iris. Lalu sama seperti tadi, kita normalisasikan terlebih dahulu kemudia kita cari vmin dan vmaxnya menggunakan perulangan for data in list_iris. Lalu untuk mencari v barunya kita bisa menggunakan rumus seperti diatas tadi dan untuk menampilkan hasilnya kita bisa panggil variabel iristesting. 

5. Prediksilah kelas data di iristesting.csv menggunakan kriteria kedekatan distance antara setiap 
data di iristesting.csv ke centroid masing2 class di iris. Gunakan euclidean distance

```python
classesiris = iris['variety'].unique()
list_iris = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

for classes in classesiris:
  distance = 0
  for data in list_iris:
    distance += (iristesting[data] - centroid[classes][data])**2
  distance = distance**(1/2)
  iristesting[classes] = distance

iristesting
```

Penjelasan : pertama kita memakai variabel iris dengan index variety lalu juga memakai .unique dan ditampung dalam variabel classesiris. Lalu sama kita ambil setiap atributnya kemudian di buat list dan ditampung dalam variabel list_iris. Lalu kita menggunakan perulangan for classes in classesiris lalu kita deklarasikan variabel distance yang bernilai nol. Lalu kita gunakan perulangan for didalam for kemudian kita buat variabel distance, nah arti variabel tersebut adalah distance itu sendiri dan kita tambahkan dengan operasi (iristesting[data] - centroid[classes][data])**2. Lalu hasil yang dapat akan diakar pangkat kemudian dimasukkan ke variabel distance. Lalu nilai dari distance itu sama dengan nilai iristesting index classes. Lalu untuk memanggil prediksinya kita panggil variabel iristesting. 

6. Hitunglah berapa persen klasifikasi yang benar (akurasi) saat memprediksi class

```python
irisresult = iristesting[classesiris]
result = irisresult = irisresult.idxmin(axis=1)

correct = (iristesting['variety'] == result).sum()

total = len(iristesting)
akurasi = (correct / total) * 100
print("akurasi: ", akurasi, "%")
```

Penjelelasan : Kita bisa memakai variabel iristesting dnegan index classesiris kemudian ditampung di variabel irisresult. irisresult.idxmin(axis=1) digunakan untuk mengembalikan nilai index minimum dengan sumbu 1 laluditampung dalam variabel result. Lalu variabel correct menampung variabel iristesting dengan index variety = result kemudian akan di sum dan ditampung dalam variabel correct tersebut. Lalu len digunakan untuk menghitung panjang seluruh variabel iristesting dan ditampung dalam variabel total. Lalu untuk menghitung akurasi kita bisa menghitung dengan cara membagi correct dengan total kemudian dikali 100 dan kita mendapatkan hasil akurasi = 96.66666666666667 % .
